```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Pastikan dataset telah dimuat
df = pd.read_csv('students.csv', delimiter=';')

# Distribusi Variabel: Visualisasi dengan histogram dan boxplot
# Histogram untuk setiap fitur numerik
df.hist(bins=20, figsize=(10, 10))
plt.suptitle("Distribusi Nilai Mahasiswa", fontsize=16)
plt.show()

# Boxplot untuk setiap fitur numerik
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include=['number']))  # Hanya numerik
plt.title("Boxplot Distribusi Nilai Mahasiswa")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Pastikan dataset telah dimuat dengan benar
df = pd.read_csv('students.csv', delimiter=';')

# Hubungan Antar Variabel: Visualisasi menggunakan pairplot dan scatter plot
# Pairplot untuk melihat hubungan antar variabel numerik
sns.pairplot(df, hue='Jenis Kelamin', markers=["o", "s"])
plt.suptitle("Pairplot Nilai Mahasiswa Berdasarkan Jenis Kelamin", y=1.02)
plt.show()

# Scatter plot untuk hubungan spesifik
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Matematika', y='IPA', hue='Jenis Kelamin', data=df, palette='viridis')
plt.title("Scatter Plot Nilai Matematika vs IPA Berdasarkan Jenis Kelamin")
plt.xlabel("Nilai Matematika")
plt.ylabel("Nilai IPA")
plt.legend(title="Jenis Kelamin")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Memuat dataset
df = pd.read_csv('students.csv', delimiter=';')

# Visualisasi Kelas: Distribusi kelas dalam dataset menggunakan count plot atau pie chart

# Count plot untuk distribusi Jenis Kelamin
plt.figure(figsize=(10, 6))
sns.countplot(x='Jenis Kelamin', data=df, palette='viridis')
plt.title('Distribusi Jenis Kelamin dalam Dataset')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah')
plt.show()

# Count plot untuk distribusi Jurusan
plt.figure(figsize=(10, 6))
sns.countplot(x='Jurusan', data=df, palette='coolwarm')
plt.title('Distribusi Jurusan dalam Dataset')
plt.xlabel('Jurusan')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()

# Pie chart untuk distribusi Jenis Kelamin
df['Jenis Kelamin'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(8, 8), colors=['skyblue', 'lightcoral'])
plt.title('Distribusi Jenis Kelamin dalam Dataset')
plt.ylabel('')
plt.show()

# Pie chart untuk distribusi Jurusan
df['Jurusan'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(8, 8), colors=['lightgreen', 'lightblue', 'salmon', 'gold'])
plt.title('Distribusi Jurusan dalam Dataset')
plt.ylabel('')
plt.show()
```
